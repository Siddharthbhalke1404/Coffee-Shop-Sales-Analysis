```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import datetime
```

```python
df=pd.read_csv('Coffee Shop Sales.csv')
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.info()
```

## Converting DataType Of Transaction Time

```python
df['transaction_time']=df['transaction_time'].astype(str)
```

```python
df['transaction_time']=pd.to_timedelta(df['transaction_time'])
```

```python
df.info()
```

## Converting DataType Of Transaction Date

```python
df['transaction_date']=pd.to_datetime(df['transaction_date'],format= '%d/%m/%Y')
```

```python
df.info()
```

## Creating New Column Of Sales

```python
df.head(1)
```

```python
df['Sales'] = df['transaction_qty'] * df['unit_price']
```

```python
df.head(1)
```

## Adding Date And Time Columns

```python
df.head(1)
```

```python
df['Datetime'] = df['transaction_date'] + df['transaction_time']
```

```python
df.head(1)
```

## Find Total Sales with respect to Locations

```python
df_location=df.groupby(df['store_location']).aggregate({
    'Sales':'sum',
    'transaction_id':'count'    
})

df_location
```

```python
df.head(1)
```

```python
df.describe()
```

## Visuals of store_location Sales

```python
#Sales by Location every day
Daily_sales_by_location=df.groupby(['transaction_date','store_location'])['Sales'].sum().unstack()
Daily_sales_by_location
```

```python
Daily_sales_by_location.plot(figsize=(10,4),title="Daily Sales By Store Location")
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend(title='Store Location')
plt.grid(True)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.stackplot(Daily_sales_by_location.index,Daily_sales_by_location.T,labels=Daily_sales_by_location.columns)
plt.title('Daily Sales By store Location (Stacked Area Chart)')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend(title='Store Location',loc='upper left')
plt.grid(True)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

#  Total Sales by per day

```python
daily_sales=df.groupby(['transaction_date'])['Sales'].sum().reset_index()
daily_sales
```

# Total Sales by Per Week

```python
df['day_of_week'] = df['Datetime'].dt.day_name()
weekly_sales = df.groupby('day_of_week')['Sales'].sum().reset_index()
```

```python
weekly_sales
```

# Total Sales by Per Month 

```python
df['month'] = df['Datetime'].dt.to_period('M')
monthly_sales = df.groupby('month')['Sales'].sum().reset_index()
```

```python
monthly_sales
```

# Total Sales per month 

```python
monthly_sales.plot(figsize=(10,4),title="Monthly Sales")
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

# Weekly sales by location

```python
df.columns
```

```python
weekly_sales = df.groupby(['day_of_week','store_location'])['Sales'].sum().unstack()
days_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekly_sales = weekly_sales.reindex(days_order)
weekly_sales
```

```python
weekly_sales[["Astoria","Hell's Kitchen","Lower Manhattan"]].plot(kind='bar',figsize=(14,6))

plt.xlabel('Day of the week',fontweight='bold')
plt.ylabel('Total Sales ($)',fontweight='bold')

plt.title('Weekly Sales by Store Location')
plt.legend(title='Store Location',loc='center')

plt.tight_layout()
plt.show()
```

## Number of sales per Products_category

```python
category=df.groupby(['product_category']).agg({
    'Sales':'sum',
    'transaction_id':'count'}).sort_values('Sales',ascending=False)
```

```python
category
```

# Product_category sales by per hour

```python
df['Hour']=df['Datetime'].dt.hour
```

```python
hourly_sales_by_product=df.groupby(['Hour','product_category'])['Sales'].sum().unstack()
```

```python
hourly_sales_by_product.plot(kind='bar',stacked=True,figsize=(14,8),colormap='viridis')

plt.xlabel('Hour Of The Day',fontweight='bold')
plt.ylabel('Total Sales ($)',fontweight='bold')

plt.title('Hourly Sales by Product Category ')
plt.legend(title='Product Category',loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Sales By Product

```python
df.columns
```

```python
Sales_by_product=df.groupby(['product_type']).agg({
    'Sales':'sum',
    'unit_price':'mean'}).sort_values('Sales',ascending=False)
```

```python
Sales_by_product
```

```python
product_sales=df.groupby(['product_category','product_type'])['Sales'].sum().reset_index()
```

```python
product_sales
```

# Sales By Products

```python
Categories = product_sales['product_category'].unique()

for i in Categories:
    plt.figure(figsize=(5,5))
    category_data = product_sales[product_sales['product_category']==i]
    sns.barplot(data=category_data,x='product_type',y='Sales') 
    
    plt.title(f'Sales by product type in {i}')
    plt.xlabel('Product Type')
    plt.ylabel('Total Sales ($)')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

# Hourly Sales By Store Location

```python
hourly_sales_by_location=df.groupby(['store_location','Hour'])['Sales'].sum().reset_index()
```

```python
hourly_sales_by_location
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data=hourly_sales_by_location,x='Hour',y='Sales',hue='store_location')

plt.title('Hourly Sales by Store Location')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales')
plt.legend(title='store_location',loc='upper right')
plt.grid(True)
plt.xticks(rotation=45)
    
plt.tight_layout()
plt.show()
```
